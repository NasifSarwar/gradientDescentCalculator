```python
w1=0.02
w2=0.10
w3=0.07
w0=0.01
alpha= 0.01
```

```python
dwtl1=0
dwtl2=0
dwtl3=0
dwtl0=0
```

```python
print(f"w1= {w1}")
print(f"w2= {w2}")
print(f"w3= {w3}")
print(f"w0= {w0}")
x1= [12, 17, 13, 14, 15]
x2= [7, 8, 10, 9, 11]
x3= [14, 11, 12, 15, 13]
y=  [73, 80, 80, 85, 89]
n=5
z = 0
```

```python
#inside range number of epoch
for i in range(5):
    print(f'Loop number {i+1}')
    # number of data row in a loop is n
    for x in range(n):
        print(f'Data From Row: {z+1}')

        ypred=(w1*x1[z])+(w2*x2[z])+(w3*x3[z])+w0
        ydeff = y[z] - ypred
        print(f'yPred{z+1} = {ypred}')
        print(f'ydeff{z+1} = {ydeff}')
        dw1= -2 * x1[z] * ydeff
        dw2= -2 * x2[z] * ydeff
        dw3= -2 * x3[z] * ydeff
        dw0= ydeff
        dwtl1 = dwtl1+dw1
        dwtl2 = dwtl2+dw2
        dwtl3 = dwtl3+dw3
        dwtl0 = dwtl0+dw0
        print(f"dw1= {dw1}")
        print(f"dw2= {dw2}")
        print(f"dw3= {dw3}")
        print(f"dw0= {dw0}")
        print()
        print()   
        if n-1 == x:
            w1= w1 - (alpha*(dwtl1/(x+1)))
            w2= w2 - (alpha*(dwtl2/(x+1)))
            w3= w3 - (alpha*(dwtl3/(x+1)))
            w0= w0 - (alpha*(dwtl0/(x+1)))
            print(f"Updatedw1= {w1}")
            print(f"Updatedw2= {w2}")
            print(f"Updatedw3= {w3}")
            print(f"Updatedw0= {w0}")
            dwtl1=0
            dwtl2=0
            dwtl3=0
            dwtl0=0
            print()
            print()

        z+=1
        if z > 4:
            z=0
```
